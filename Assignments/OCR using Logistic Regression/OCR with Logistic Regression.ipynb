```python
import numpy as np
import pandas as pd
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def calculate_accuracy(predictions, actuals):
    return np.mean(predictions == actuals) * 100
```

```python
def logistic_regression(eta, max_iterations=1000):
    train_X = np.loadtxt('trainX.txt', dtype=np.uint8)
    col_of_ones_train = np.ones((train_X.shape[0]), dtype=np.uint8)
    train_X = np.insert(train_X, 0, col_of_ones_train, axis=1)

    train_Y = np.loadtxt('trainY.txt', dtype=np.uint8)
    train_Y = np.where(train_Y == 2, 1, 0)

    test_X = np.loadtxt('testX.txt', dtype=np.uint8)
    col_of_ones_test = np.ones((test_X.shape[0]), dtype=np.uint8)
    test_X = np.insert(test_X, 0, col_of_ones_test, axis=1)

    test_Y = np.loadtxt('testY.txt', dtype=np.uint8)
    test_Y = np.where(test_Y == 2, 1, 0)

    thetas = np.random.uniform(0, 1, size=(train_X.shape[1]))

    for iteration in range(1, max_iterations + 1):
        probabilities = sigmoid(np.dot(train_X, thetas))
        predictions = np.where(probabilities >= 0.5, 1, 0)
        thetas -= (eta / train_X.shape[0]) * np.dot(train_X.T, (probabilities - train_Y))

        accuracy = calculate_accuracy(predictions, train_Y)
        print(f'Accuracy for iteration {iteration}: {accuracy}')

        if accuracy == 100:
            break

    print(f'\nFor Eta: {eta}')
    print(f'Total number of iterations: {iteration}')
    
    train_predictions = np.where(sigmoid(np.dot(train_X, thetas)) >= 0.5, 1, 0)
    print(f'Accuracy on Training data: {calculate_accuracy(train_predictions, train_Y)}')
    
    test_predictions = np.where(sigmoid(np.dot(test_X, thetas)) >= 0.5, 1, 0)
    print(f'Accuracy on Testing data: {calculate_accuracy(test_predictions, test_Y)}\n')

etas = [0.01,0.001, 0.1, 0.3, 0.5]

for eta in etas:
    logistic_regression(eta)
```
