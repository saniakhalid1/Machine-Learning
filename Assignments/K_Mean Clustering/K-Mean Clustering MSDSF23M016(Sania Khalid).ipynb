# Image-1: Greenery Image

```python
# ----------------LIBERARIES---------------------
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# -------------Load image---------------------------
input_image = "greenery.JPG"
image = Image.open(input_image)
```

```python
#------- Conversion to numpy array------------------
image_data = np.array(image)
rows, cols, RGB = image_data.shape
image_data.shape
```

```python
# ----------Reshape the image to a 2D array-------------
reshaped_data = image_data.reshape(-1, RGB)
reshaped_data.shape
```

```python
# ---------# K-means algorithm----------------------------
def initialize_centroids(k, data):
    centroids = data[np.random.choice(data.shape[0], k, replace=False)]
    return centroids

#data[:, np.newaxis, :]: Reshapes the data to have an additional axis, making it compatible for element-wise 
#subtraction.  #distances = np.linalg.norm(data[:, np.newaxis, :] - centroids, axis=2): Calculates the Euclidean
#distances between each data point and all centroids. #labels = np.argmin(distances, axis=1): Assigns 
#each data point to the centroid with the minimum distance.#Returns an array of cluster labels indicating which centroid each data point belongs to.


def assign_points(data, centroids):
    distances = np.linalg.norm(data[:, np.newaxis, :] - centroids, axis=2)
    labels = np.argmin(distances, axis=1)
    return labels

# This line calculates the Euclidean distances between each data point and all centroids.
#The np.newaxis is used to add a new axis to the data array

def update_centroids(data, labels, k):
    centroids = np.array([data[labels == i].mean(axis=0) for i in range(k)])
    return centroids

#[data[labels == i].mean(axis=0) for i in range(k)]: Calculates the mean position of data points for each cluster.
#centroids = np.array(...): Combines mean positions into an array, representing the updated centroid positions.
#Returns the updated centroids.

def k_means(data, k, max_iters=50):
    centroids = initialize_centroids(k, data)
    for _ in range(max_iters):
        old_centroids = centroids
        labels = assign_points(data, centroids)
        centroids = update_centroids(data, labels, k)
        if np.all(old_centroids == centroids):
            break
    return centroids, labels
```

```python
#--------------- implementation for k=2, k=3, k=4-----------
k_values = [2, 3]
```

```python
input_K = int(input("enter the values of K "))
arr = input_K
```

```python
plt.figure(figsize=(15, 5))
```

```python
# ------------Original Image---------------------------------
plt.subplot(1, 4, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')

centroids, labels = k_means(reshaped_data, input_K)
clustered_data = centroids[labels].reshape(rows, cols, RGB).astype(np.uint8)

#for i, k in enumerate(k_values, start=2):
    #centroids, labels = k_means(reshaped_data, k)
    #clustered_data = centroids[labels].reshape(rows, cols, RGB).astype(np.uint8)

   
plt.subplot(1, 4, input_K)
plt.imshow(clustered_data)
plt.title(f"k={input_K}")
plt.axis('off')

plt.tight_layout()
plt.show()
```

It shows that for k=2 algorithm only shows 2 colors , for k=3 there are one added feature and for k=10 image is almost same to original image so we can easily say that for k = 3 it is compressed version of original image with less quality loss.

# Image-2: Face Image

```python
# ----------------LIBERARIES---------------------
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# -------------Load image---------------------------
input_image = "female.jpg"
image = Image.open(input_image)
```

```python
#------- Conversion to numpy array------------------
image_data = np.array(image)
rows, cols, RGB = image_data.shape
image_data.shape
```

```python
# ----------Reshape the image to a 2D array-------------
reshaped_data = image_data.reshape(-1, RGB)
reshaped_data.shape
```

```python
# ---------# K-means algorithm----------------------------
def initialize_centroids(k, data):
    centroids = data[np.random.choice(data.shape[0], k, replace=False)]
    return centroids

def assign_points(data, centroids):
    distances = np.linalg.norm(data[:, np.newaxis, :] - centroids, axis=2)
    labels = np.argmin(distances, axis=1)
    return labels

def update_centroids(data, labels, k):
    centroids = np.array([data[labels == i].mean(axis=0) for i in range(k)])
    return centroids

def k_means(data, k, max_iters=50):
    centroids = initialize_centroids(k, data)
    for _ in range(max_iters):
        old_centroids = centroids
        labels = assign_points(data, centroids)
        centroids = update_centroids(data, labels, k)
        if np.all(old_centroids == centroids):
            break
    return centroids, labels
```

```python
#--------------- implementation for k=2, k=3, k=4-----------
k_values = [2, 3, 10]
```

```python
plt.figure(figsize=(15, 5))
```

```python
# ------------Original Image---------------------------------
plt.subplot(1, 4, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
for i, k in enumerate(k_values, start=2):
    centroids, labels = k_means(reshaped_data, k)
    clustered_data = centroids[labels].reshape(rows, cols, RGB).astype(np.uint8)

    plt.subplot(1, 4, i)
    plt.imshow(clustered_data)
    plt.title(f"k={k}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# COMMENTS....

It shows that for k=2 algorithm only shows 2 colors , for k=3 there are one added feature and for k=10 image is little bit closer to original but still some features are missing . we could say that for k=10 it is compressed version of original image.
